# Create Random Dates

```python
# Set Start Date, End date
start_date = datetime.date(2022, 1, 1)
end_date = datetime.date(2022, 12, 31)

# Create Random Dates Generation Function
def random_date(start, end):
    generate_random =  start + datetime.timedelta(days=random.randint(0, (end - start).days))
    return generate_random

# Create Random Dates
random_dates = [random_date(start_date, end_date) for _ in range(1000)]

# Replace a Random Number of Elements with Null
num_to_replace = random.randint(0, 50)  # Select a Random Number from 0 to 50
indices_to_replace = random.sample(range(1000), num_to_replace)

for index in indices_to_replace:
    random_dates[index] = None

for date in random_dates:
    print(date)
```

```python
# Convert List to a DataFrame
# "cart_date" is the date the customer added an item to cart.

df = pd.DataFrame(random_dates, columns=["cart_date"])
df
```

```python
df.isna().sum()
```

```python
from datetime import datetime, timedelta


# Create Random Dates Generation Function with a Timedelta
def generate_random_date(start_date, end_date):
    random_days = random.randint(0, (end_date - start_date).days)
    return start_date + timedelta(days=random_days)

# Create Random Dates by Adding a New Column
new_dates = []
for index, row in df.iterrows():
    if pd.notna(row[0]) and row[0] != 'None':  # Not NaN and Not None
        current_date = datetime.strptime(str(row[0]), '%Y-%m-%d').date()
        end_date = datetime.strptime("2022-12-31", '%Y-%m-%d').date()
        random_date = generate_random_date(current_date, end_date)
        new_dates.append(random_date.strftime('%Y-%m-%d'))
        
    else:
        new_dates.append(None)

# Add a New Column to a DataFrame
# "buy_date" is the date the customer purchased the item in their cart
df['buy_date'] = new_dates

# Print Result
df
```

# Save DataFrame as a File

```python
# Set File Name
file_name = "date.xlsx"

# Save the DataFrame as an XLSX File
df.to_excel(file_name, index=False)
```
